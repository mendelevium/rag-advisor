```python
import json
from dotenv import load_dotenv
load_dotenv()
```

```python
import openai

def generate_response(messages):
    response = openai.chat.completions.create(
        model="gpt-4o-mini",
        messages=messages,
        max_tokens=256,
        stream=False
    )
    return response.choices[0].message.content
```

```python
with open('../data/6023Ecouteurs.json', 'r', encoding='utf-8') as file:
    data = json.load(file)

print(data)
```

```python
prompt = "Provide a french summary of the following product details:\n\n" + json.dumps(data[1], indent=2)
messages = [
    {"role": "system", "content": "You are a helpful assistant that summarizes product details in french. Don't use markdown. Don't use html. Do not use any formatting. Just plain text. Don't mention where to find it or how to buy it. Just provide a summary of the product details."},
    {"role": "user", "content": prompt}
]
summary = generate_response(messages)
print(summary)
```

```python
sum_list = []
for i in range(1, len(data)):
    prompt = "Provide a french summary of the following product details:\n\n" + json.dumps(data[i], indent=2)
    messages = [
        {"role": "system", "content": "You are a helpful assistant that summarizes product details in french. Don't use markdown. Don't use html. Do not use any formatting. Just plain text. Don't mention where to find it or how to buy it. Just provide a summary of the product details."},
        {"role": "user", "content": prompt}
    ]
    summary = generate_response(messages)
    sum_list.append(summary)
    print(summary)
    # print(data[i]['properties'])
```

```python
with open('../data/summaries.txt', 'w', encoding='utf-8') as f:
    f.write(str(sum_list))
```
